1. You are given a data set, which is present in the LMS, containing the number of hurricanes occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

h = pd.read_csv("Hurricanes.csv")
df = pd.DataFrame(h)

#df.plot(x='Year',y='Hurricanes',kind='bar')

plt.bar(df['Year'], df['Hurricanes'])
plt.xticks(np.arange(min(df['Year']),max(df['Year'])+1,10),rotation=70)
plt.xlabel('Years')
plt.ylabel('Number of hurricanes')
plt.title('Hurricanes by year')
plt.show()
```

3. Plot a piechart of the number of models released by every manufacturer, recorded in the data provided. Also mention the name of the manufacturer with the largest releases.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

m = pd.read_csv('Cars2015.csv')
df = pd.DataFrame(m)
df['Make'] = df['Make'].str.strip()

df2 = df.groupby(['Make'])['Model'].count()
#df2.plot(kind='pie')
#df2.max('Model')

explode = df2.copy()
explode.values[:]=0
explode[df2.idxmax()]=.2
print("Manufacturer who has maximum number of models is: ",df2.idxmax())


cmap = plt.get_cmap('Spectral')
clist = [cmap(i) for i in np.linspace(0,1,len(set(df['Make'])))]
plt.figure(figsize=(12,10))
plt.pie(df2,labels=df2.index,colors=clist,explode=explode)
#plt.title("Pie chart for Makes number of models")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as datetime

df = pd.read_csv('BigMartSalesData.csv')
```

```python
#1. Plot total sales per month for the Year 2011. How have the total sales
#increased over the months in the Year 2011? Which month has the lowest
#sales?

dfg2011m = df[df['Year']==2011].groupby(['Month'])['Amount'].sum()

Month_names=list()


clist = dfg2011m.astype(str)
clist.values[:]='b'
clist[dfg2011m.idxmin()]='r'

#plt.bar(dfg2011m.index,dfg2011m, tick_label= Month_names)
plt.bar(dfg2011m.index,dfg2011m, 
        tick_label= [(datetime.datetime.strptime(str(i), "%m")).strftime("%B") for i in np.arange(1,len(dfg2011m)+1)],
       color=clist)
plt.xticks(rotation=90)
plt.xlabel('Months')
plt.ylabel('Amount')
plt.title("Year 2011 sales by month")
plt.show()
```

```python
#2. Plot a Pie Chart for the Year 2011 Country Wise. Which Country contributes
#highest towards sales?

dfg2011c = df[df['Year']==2011].groupby(['Country'])['Amount'].sum()

explodelist = dfg2011c.copy()
explodelist.values[:]=0
explodelist[dfg2011c.idxmax()]=.5


cmap = plt.get_cmap('Spectral')
clist = [cmap(i) for i in np.linspace(0,1,len(dfg2011c))]

plt.figure(figsize=(10,8))
plt.pie(dfg2011c,labels=dfg2011c.index,colors=clist, explode=explodelist,shadow=True)
plt.rcParams.update({'font.size':8})
plt.show()
```
